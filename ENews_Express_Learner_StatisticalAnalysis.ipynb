## 1. Do the users spend more time on the new landing page than the existing landing page?

### Perform Visual Analysis

```python
# visual analysis of the time spent on the new page and the time spent on the old page
plt.figure(figsize=(10,8))
sns.boxplot(x = 'landing_page', y = 'time_spent_on_the_page', data = df)
plt.show()
```

### Step 1: Define the null and alternate hypotheses

$H_0$: Users spend the same or less amount of time on the new landing page compared to the existing page.

$H_a$: Users spend more time on the new landing page

### Step 2: Select Appropriate test

We use one-tailed test because the alternative hypothesis is about more time spent on the new landing page.

### Step 3: Decide the significance level

significance level of 5% is provided in the problem description. $\alpha = 0.05$.

### Step 4: Collect and prepare data

```python
time_spent_new = df[df['landing_page'] == 'new']['time_spent_on_the_page']
time_spent_old = df[df['landing_page'] == 'old']['time_spent_on_the_page']
print('Sample standard deviation of the time spent on the new page is:', time_spent_new.std())
print('Sample standard deviation of the time spent on the old page is:',time_spent_old.std())
```

### Step 5: Calculate the p-value

```python
from scipy.stats import stats
# Calculating p-value
t_statistic, p_value =  stats.ttest_ind(time_spent_new, time_spent_old, equal_var = True, alternative = 'greater')
p_value
```

### Step 6: Compare the p-value with $\alpha$

```python
# p-value comparison with alpha
if p_value < 0.05:
    print(f'The p-value is less than the level of significance, we reject the null hypothesis.')
else:
    print(f'The p-value is greater than the level of significance, we fail to reject the null hypothesis.')
```

### Step 7:  Draw inference

As we reject the null hypothesis, we have enough statistical evidence to say that the users spend more time on the new landing page than the existing landing page

**A similar approach can be followed to answer the other questions.**

## 2. Is the conversion rate (the proportion of users who visit the landing page and get converted) for the new page greater than the conversion rate for the old page?

### Visual Analysis

```python
plt.figure(figsize=(10,8))
sns.boxplot(x = 'converted', y = 'landing_page', data = df)
plt.show()
```

```python
crosstab_landing_page= pd.crosstab(df['converted'], df['landing_page'])
print(crosstab_landing_page)
```

```python
# conversion rate for the new page vs the old page
pd.crosstab(df['converted'],df['landing_page'],normalize='index').plot(kind="bar", figsize=(6,8),stacked=True)
plt.legend()
plt.show()
```

### Step 1: Define the null and alternate hypotheses

$H_0$: The conversion rate for the new page is less than or equal to that of the old page.

$H_a$: The conversion rate for the new page is greater than that of the old page.

### Step 2: Select Appropriate test

Since we are checking whether the new page has greater conversion rate, this will be **a one-tailed test**.

### Step 3: Decide the significance level

significance level of 5% is provided in the problem description. $\alpha = 0.05$.

### Step 4: Collect and prepare data

```python
converted_treatment_new = df[df['group'] == 'treatment']['converted'].value_counts()['yes']
converted_control_old = df[df['group'] == 'control']['converted'].value_counts()['yes']
total_users_control = df.group.value_counts()['control']
total_users_treatment = df.group.value_counts()['treatment']
print('The numbers of users served the new and old pages are {0} and {1} respectively'.format(total_users_control, total_users_treatment ))
```

### Step 5: Calculate the p-value

```python
from scipy.stats import stats
t_statistic, p_value = stats.ttest_ind([converted_treatment_new, converted_control_old] , [total_users_treatment, total_users_control], alternative ='less')
p_value
```

### Step 6: Compare the p-value with $\alpha$

```python
if p_value < 0.05:
    print(f'The p-value is less than the level of significance, we reject the null hypothesis.')
else:
    print(f'The p-value is greater than the level of significance, we fail to reject the null hypothesis.')
```

### Step 7:  Draw inference

As we reject the null hypothesis, we have enough statistical evidence to say the conversion rate for the new landing page is greater than the conversion rate for the old landing page.

## 3. Is the conversion and preferred language are independent or related?

### Perform Visual Analysis

```python
# The dependency between conversion status and preferred langauge
pd.crosstab(df['converted'],df['language_preferred'],normalize='index').plot(kind="bar", figsize=(6,8))
plt.legend()
plt.show()
```

### Step 1: Define the null and alternate hypotheses

$H_0:$ Conversion is independent of Preferred Language.

$H_a:$ Conversion is dependent of Preferred Language.




### Step 2: Select Appropriate test

For the test of independence, concerning two categorical variables - converted status and preferred language, we use **chi-square test of independence**.

### Step 3: Decide the significance level

As given in the problem statement, we select **α = 0.05.**

### Step 4: Collect and prepare data

```python
# complete the code to create a contingency table showing the distribution of the two categorical variables
contingency_table = pd.crosstab(df['converted'], df['language_preferred'])
contingency_table
```

### Step 5: Calculate the p-value

```python
from scipy.stats import chi2_contingency
chi2, p_value, dof, exp_freq = chi2_contingency(contingency_table)
print('The p-value is', p_value)
```

### Step 6: Compare the p-value with $\alpha$

```python
if p_value < 0.05:
    print(f'The p-value is less than the level of significance, we reject the null hypothesis.')
else:
    print(f'The p-value is greater than the level of significance, we fail to reject the null hypothesis.')
```

### Step 7:  Draw inference

As the above statistical evidence fail to reject the null hypothesis, we conclude that **Conversion & Preferred Language are independant**.


## 4. Is the time spent on the new page same for the different language users?

### Perform Visual Analysis

```python
# users who got served the new page
df_new = df[df['landing_page'] == 'new']
```

```python
# The time spent on the new page for different language users
plt.figure(figsize=(8,8))
sns.boxplot(x = 'language_preferred', y = 'time_spent_on_the_page', data = df_new)
plt.show()
```

```python
# Mean time spent on the new page for different language users
df_new.groupby(['language_preferred'])['time_spent_on_the_page'].mean()
```

### Step 1: Define the null and alternate hypotheses

$H_0:$ The time spent on the new page **same** for the different language users

$H_a:$ The time spent on the new page is **different** for different language users



### Step 2: Select Appropriate test

**ANOVA (Analysis of Variance) is used to compare the mean of multiple groups.**

### Step 3: Decide the significance level

As given in the problem description, we select α = 0.05.

### Step 4: Collect and prepare data

```python
# Time spent on the new page by English, French and Spanish language users
time_spent_English = df_new[df_new['language_preferred']=="English"]['time_spent_on_the_page']
time_spent_French = df_new[df_new['language_preferred']=="French"]['time_spent_on_the_page']
time_spent_Spanish = df_new[df_new['language_preferred']=="Spanish"]['time_spent_on_the_page']
```

### Step 5: Calculate the p-value

```python
from scipy.stats import stats
# calculating p-value
test_stat, p_value = stats.f_oneway(time_spent_English, time_spent_French, time_spent_Spanish)
print('The p-value is', p_value)
```

### Step 6: Compare the p-value with $\alpha$

```python
if p_value < 0.05:
    print(f'The p-value is less than the level of significance, we reject the null hypothesis.')
else:
    print(f'The p-value is greater than the level of significance, we fail to reject the null hypothesis.')
```

### Step 7:  Draw inference

As the above statistical evidence fail to reject the null hypothesis, we conclude that the time spent on the new page is different for different language users.
